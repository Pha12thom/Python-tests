```python
def selection_sort(items):
    for i in range(0,len(items)-1):
        for j in range(i+1,len(items)):
            if items[i] > items[j]:
                items[i], items[j] = items[j], items[i]
    return items
print(selection_sort([1,5,6,3,2,8,9,5,3]))
```
